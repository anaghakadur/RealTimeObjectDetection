```python
!python detect.py --weights runs/train/aquarium_yolov5/weights/best.pt --img 416 --source /kaggle/working/aquarium-data-cots/aquarium_pretrain/test/images --save-txt --save-conf --name infer_aquarium
```

```python

from ultralytics import YOLO

model = YOLO("yolov8s.pt")  # or yolov8n.pt for lightweight version

model.train(
    data=yaml_path,
    epochs=30,
    imgsz=416,
    batch=16,
    name="aquarium_yolov8"
)
```

```python
metrics = model.val()
print(metrics)
```

```python
model.predict(
    source="/kaggle/input/aquarium-data-cots/aquarium_pretrain/test/images",
    save=True,
    save_txt=True,
    name="aquarium_preds_yolov8"
)
```
